```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
# Loading Dataset
df = pd.read_csv('Booking Datasets.csv')
```

```python
# Converting the 'reservation_status_date' column to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True)
```

```python
# Viewing the description of columns with object data types
df.describe(include='object').T
```

```python
# Checking unique values in each object column
for col in df.describe(include = 'object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
# Checking for null values in the dataset
df.isnull().sum()
```

```python
# Checking for null values in the dataset
df.isnull().sum()
```

```python
# Dropping the 'company' and 'agent' columns 
df.drop(['company','agent'], axis = 1, inplace = True) 
# Dropping rows with null values
df.dropna(inplace = True)
```

```python
# Rechecking for null values in the dataset after dropping columns and rows
df.isnull().sum()
```

```python
# Checking the statistical summary of the dataset
df.describe().round(2).T
```

```python
# Filtering out rows where 'adr' is greater than or equal to 5000
df = df[df['adr']<5000]
```

```python
#Data Analysis and Visualisation
cancelled_perc = df['is_canceled'].value_counts(normalize = True)    
print(cancelled_perc)                                                 # 63% bookings were not cancelled
plt.figure(figsize = (5, 4))
plt.title('Reservation Status Count')
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7,
        color=['#4CAF50', '#F44336'])
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette=['#4CAF50','#F44336'])
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in Different Hotels', size = 15)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.legend(['Not Cancelled','Cancelled'])
plt.show()
```

```python
# Checking cancellation percentage for Resort Hotel
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True) * 100
```

```python
# Checking cancellation percentage for City Hotel
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True) *100
```

```python
# Grouping the data by 'reservation_status_date' and calculating the mean of 'adr' for both Resort Hotel and City Hotel
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 25)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel', color = '#4CAF50')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel', color = '#F44336')
plt.legend(fontsize= 20)
plt.show()
```

```python
df['month_name'] = df['reservation_status_date'].dt.month_name()
plt.figure(figsize = (16,8))
sns.countplot(
    x='month_name',
    hue='is_canceled',
    data=df,
    order=['January','February','March','April',
           'May','June','July','August',
           'September','October','November','December'] , 
           palette=['#4CAF50','#F44336']
)
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status Per Month', size = 20)
plt.xlabel('Months')
plt.ylabel('Number of Reservation')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
plt.figure(figsize=(15,8))

sns.barplot(
    x='month_name',
    y='adr',
    data=df[df['is_canceled'] == 1]
           .groupby('month_name')[['adr']]
           .sum()
           .reset_index(),
    order=['January','February','March','April',
           'May','June','July','August',
           'September','October','November','December'],
    palette='Reds'
)

plt.title('ADR Per Month (Cancelled Reservations)', fontsize=20)
plt.xlabel('Months')
plt.ylabel('Total ADR')

plt.xticks(rotation=45)
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]

top_10_country = cancelled_data['country'].value_counts()[:10]
total = top_10_country.sum()

plt.figure(figsize=(10,6))

ax = sns.barplot(
    x=top_10_country.values,
    y=top_10_country.index,
    palette='Reds_r'
)

for i, v in enumerate(top_10_country.values):
    percentage = (v / total) * 100
    plt.text(v + 50, i, f'{percentage:.2f}%', va='center')

plt.title('Top 10 Countries with Cancelled Reservations')
plt.xlabel('Number of Cancelled Reservations')
plt.ylabel('Countries')

plt.show()
```

```python

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean() 
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date',inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'Not Cancelled', color='#4CAF50')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'Cancelled', color='#FF5722')
plt.legend()
```
